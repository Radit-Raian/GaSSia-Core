```python
import io
import os
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from astroquery.ipac.ned import Ned
import astropy.units as u
```

```python
with open("host_galaxy_ascii.txt", "r") as f:
    lines = f.readlines()

start_idx = 0
for i, line in enumerate(lines):
    if "2004ef" in line:
        start_idx = i
        break

df = pd.read_csv(io.StringIO("".join(lines[start_idx:])), 
                 sep=r'\s+', 
                 header=None, 
                 engine='python')


def reconstruct_galaxy(row):
    parts = []
    for item in row[1:]:
        if str(item).lower() in ['sa', 'sb', 's', 's0', 'e', 'cd', 'cdots']:
            break
        parts.append(str(item))
    return " ".join(parts)

df['Full_Galaxy_Name'] = df.apply(reconstruct_galaxy, axis=1)

def clean_mag(val):
    if val == 'cdots' or pd.isna(val):
        return np.nan
    return float(str(val).split('(')[0])

print("Successfully extracted Galaxy Names:")
print(df[['Full_Galaxy_Name']])

df[['Full_Galaxy_Name']].to_csv("data/extracted_galaxies.csv", index=False)
```

```python
folder_path = "data/spectra"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
    print(f"Created directory: {folder_path}")

df_names = pd.read_csv('data/extracted_galaxies.csv')
galaxies = df_names['Full_Galaxy_Name'].unique()

def download_all_spectra(galaxy_list):
    for name in galaxy_list:
        print(f"\n--- Processing: {name} ---")
        try:
            spectra_list = Ned.get_spectra(name)
            
            if spectra_list:
                print(f"Found {len(spectra_list)} spectra for {name}")
                
                for i, spec in enumerate(spectra_list):
                    clean_name = name.replace(" ", "_").replace("+", "plus")
                    file_name = f"{folder_path}/{clean_name}_spec_{i}.fits"
                    
                    spec.writeto(file_name, overwrite=True)
                    print(f"Saved: {file_name}")
            else:
                print(f"No digital spectra available in NED for {name}")
                
        except Exception as e:
            print(f"Could not retrieve data for {name}: {e}")
        
        time.sleep(1.5)

download_all_spectra(galaxies)
```
