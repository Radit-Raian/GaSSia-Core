```python
from astropy.io import fits
import numpy as np

from astropy.io import fits
import numpy as np

hdul = fits.open("data/spectra/ESO_107-G4_NED.fits")
data = hdul[0].data

flux = data[0]
wave = data[1]

mask = np.isfinite(wave) & np.isfinite(flux)

wave = wave[mask]
flux = flux[mask]

c = 299792.458  # km/s

log_wave = np.log10(wave)

dlog = np.mean(np.diff(log_wave))
```

```python
from ppxf.ppxf_util import log_rebin
c = 299792.458  # km/s

# compute log wavelength step
logLam = np.log(wave)
dlog = np.mean(np.diff(logLam))

velscale = c * dlog

flux_log, logLam, velscale = log_rebin(
    [wave.min(), wave.max()],
    flux,
    velscale=velscale
)
```
